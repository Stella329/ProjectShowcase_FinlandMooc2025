**Developing a larger programming project**

This fourth part culminates in a slightly larger programming project.

```python
# my ans

def main():
    user_inputs = user_input() # list[string1, string2]
    # print(user_inputs)
    scorce_result(user_inputs)

def user_input():
    input_list = []
    while True:
        inputs = input('Exam points and exercises completed:')
        if inputs == "":
            break
        input_list.append(inputs.split(" ")) # list[sub_list] --[['10', '20'], ['30', '40']]; string
    return input_list

def grade(score:float):
    '''为输入score判定grade'''
    grade = 0
    if 0 <= score <= 14:
        grade = 0
    elif 15 <= score <= 17:
        grade = 1
    elif 18 <= score <= 20:
        grade = 2
    elif 21 <= score <= 23:
        grade = 3
    elif 24 <= score <= 27:
        grade = 4
    else:
        grade = 5
    return grade

def scorce_result(List:list): 
    '''转化输入string为list，计算每次总score，并判定exam_point<10特殊情况'''
    # 最后总分
    score_list = []
    # grade星星
    grade_list = ["","","","","",""] # index = grade-1 -》012345
    # 多少pass了
    pass_count = 0

    # 计算分数
    for sub_list in List:
        exam_points = int(sub_list[0])

        exercise = int(sub_list[1])
        # 要求round down: The number of exercise points granted is an integer value, rounded down.
        exercise_points = int(exercise * 0.1) # 截断取整

        # 最终总分
        sum_score = exam_points + exercise_points
        score_list.append(sum_score)
        # print(f'score = {score_list}')

        '''调用grade()判定分数'''
        g = grade(sum_score) # 调用grade()函数：计算score的grade
        # print(g)

        '''cutoff threshold： 不及格特殊要求'''
        if exam_points >= 10 and g != 0: 
            pass_count +=1 # 题目要求：f a student received less than 10 points from the exam, they automatically fail the course

            '''画星星'''
            g = grade(sum_score) # 调用grade()函数：计算score的grade
            # print(g)
            grade_list[g]+= "*"  # index：画星
        
        else:
            grade_list[0]+= "*" # 0分档

    #平均分
    avg = sum(score_list) / len(score_list)

    # 打印最终结果
    print('Statistics:')
    print(f'Points average: {avg:.1f}')
    pass_percentage = pass_count/len(score_list)
    print(f'Pass percentage: {pass_percentage * 100:.1f}')

    # print grade distribution
    print('Grade distribution:')
    for num in range(len(grade_list)-1,-1,-1): # 5,4,3,2,1,0 分别打印每个等级(reversed)
        print(f'{num}: {grade_list[num]}')

main()
```
