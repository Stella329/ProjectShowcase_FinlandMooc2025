**REQUEST:**

The file lottery_numbers.csv containts winning lottery numbers in the following format:

Sample data

week 1;5,7,11,13,23,24,30

week 2;9,13,14,24,34,35,37

...etc...

---

Each line should contain a header week x, followed by seven integer numbers which are all between 1 and 39 inclusive.

The file has been corrupted. Lines in the file may contain the following kinds of errors (these exact lines may not be present in the file, but errors in a similar format will be):

The week number is incorrect:


week zzc;1,5,13,22,24,25,26

One or more numbers are not correct:


week 22;1,**,5,6,13,2b,34

Too few numbers:


week 13;4,6,17,19,24,33

The numbers are too small or large:


week 39;5,9,15,35,39,41,105

The same number appears twice:


week 41;5,12,3,35,12,14,36

---

Please write a function named filter_incorrect(), which creates a file called correct_numbers.csv. The file should contain only those lines from the original file which are in the correct format.

```python
# My Code

def corrupted_table(file_name:str):
    with open (file_name) as origin_file:
        corrupted_table = {}
        for line in origin_file:
            line = line.strip()
            week = line.split(';')[0]
            nums = line.split(';')[1] #--string
            corrupted_table[week] = nums
        # print(f'corrupted_table={corrupted_table}')
        return corrupted_table

def write_csv(data:dict):
    with open('06_19_correct_numbers.csv', 'w') as file:
        for key, value in data.items():
            # map(str, Thelist) → 转成字符串
            # ", ".join(Alist) → 拼接成用逗号+空格分隔的字符串
            numbers = ','.join(map(str, value))
            file.write(f'{key};{numbers}\n')


def filter_incorrect():
    table = {}

    rough_table = corrupted_table('06_19_lottery_numbers.csv')
    for key,value in rough_table.items():
        week = key
        rough_data = value.split(',') #--list
        
        try: 
            week_num = int(week.split(' ')[1]) # Q1: The week number is incorrect -week zzc
        except ValueError:
            continue  # 跳过此次loop --删除错误数据
        # print(f'week_num={week_num}')

        try: 
            if len(rough_data) < 7: # Q2 Too few numbers
                raise ValueError ("Too few numbers")
        except ValueError:
            continue
        # print(f'week={week}')


        data = []
        try:
            for num in rough_data: # num ->string
                if not 1<= int(num) <= 39:
                    raise ValueError('The numbers are too small or large') #Q4
                elif int(num) in data:
                    raise ValueError('The same number appears twice') # Q5
                data.append(int(num)) # Q3 incorrect num: 1,**,5,6
                # print(f'data = {data}')
        except ValueError:
            continue
        # print(f'data = {data}')

        table[week] = data

    write_csv(table)
    return table


if __name__ == "__main__":
    # print(filter_incorrect())
    print(len(filter_incorrect()))
```
